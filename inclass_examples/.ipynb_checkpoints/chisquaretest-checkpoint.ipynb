```python
# import packages
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#We're going to make time series drawn from the standardized normal distribution.  
#One of these will be the population and the sample we will multiple by 1.5, which will make the 
#sample's standard deviation 1.5 rather than 1.
#Then, we're going to test to see if a chi-square test can detect that the variance is different

#population
n_pop=1000
x1=np.random.normal(0,1,size = (n_pop))

scale_factor=1.5
n_samp=60
x2=scale_factor*np.random.normal(0,1,size=(n_samp))

f=plt.figure()
plt.plot(x1)
plt.xlabel('time')
plt.ylabel('value')
plt.title('population')

f=plt.figure()
plt.plot(x2)
plt.xlabel('time')
plt.ylabel('value')
plt.title('sample')
```

```python
#calculate the population and sample standard deviations
x1_std=np.std(x1) #population
x2_std=np.std(x2) #sample

#calculate degrees of freedom for the sample
nu=n_samp-1

#calculate chi2:
chi2=(n_samp-1)*x2_std**2/x1_std**2

#calculate critical chi2 value
alpha=0.05
chi2_crit_upper=stats.chi2.ppf(1-0.05/2,df=nu)
chi2_crit_lower=stats.chi2.ppf(0.05/2,df=nu)

print('chi-square at '+str(alpha/2*100)+' percentile:',chi2_crit_lower)
print('chi-square at '+str((1-alpha/2)*100)+' percentile:',chi2_crit_upper)
print('chi-square of s^2/\sigma^2:',chi2)

if chi2<chi2_crit_lower or chi2>chi2_crit_upper: print('sample variance significantly different')
else: print('sample variance statistically consistent with population variance')
```
