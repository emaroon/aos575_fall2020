```python
##### ATOC7500 - variance example - Lecture #1
##### Originally coded by Prof. Barnes (CSU), updated/adapted for Jupyter Notebook/Python3 by Prof. Kay (CU)
##### updated/adapted for AOS 575 by Prof. Maroon (UWisc)
##### Last updated: Sept. 7, 2020
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # library for data analysis for text files (everything but netcdf files)
```

```python
## read in the data
filename = 'truax_2019.csv'
data=pd.read_csv(filename,sep=',')
X=data['airtemp_degc']
```

```python
## Plot hourly temperature data
xindex = (np.arange(0,np.size(X),1))/24.;
plt.plot(xindex, X);
v = np.std(X);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^{\circ}$C', ha='right', transform=plt.gca().transAxes, color = 'k');
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('day');
plt.title('Hourly temperature at Truax Field');
```

```python
### Plot hourly temperature data sampled once a day
plt.plot(xindex[::24],X[::24], color = 'C1');
v = np.std(X[::24]);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^{\circ}$C', ha='right', transform=plt.gca().transAxes, color = 'C1');
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled once-a-day');
```

```python
### Plot hourly temperature data sampled once a month
plt.plot(xindex[::720],X[::720], color = 'k');
v = np.std(X[::720]);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^{\circ}$C', ha='right', transform=plt.gca().transAxes, color = 'k');
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled once-a-month');
```

```python
###Plotting all on same plot
plt.plot(xindex, X,label='hourly');
plt.plot(xindex[::24],X[::24], color = 'C1',label='daily');
plt.plot(xindex[::720],X[::720], color = 'k',label='monthly');
plt.legend()
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled once-a-month');
```

```python
### Plot sampled daily over one month
plt.plot(xindex[4000:4000+720:24],X[4000:4000+720:24], color = 'C2');
v = np.std(X[4000:4000+720:24]);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^{\circ}$C', ha='right', transform=plt.gca().transAxes, color = 'C2');
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled daily over one month');
```

```python
### Sampled hourly over one day
plt.plot(xindex[4000:4000+24:1],X[4000:4000+24:1], color = 'C3');
v = np.std(X[4000:4000+24:1]);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^{\circ}$C', ha='right', transform=plt.gca().transAxes, color = 'C3');
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled hourly over one day');
```
