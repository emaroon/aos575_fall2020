```python
##import packages
import numpy as np                    #for numeric computing
import matplotlib.pyplot as plt       #for plotting
import pandas as pd                   #library for data analysis for text files (everything but netcdf files)
```

```python
## read in the data
filename = 'truax_2019.csv'
data=pd.read_csv(filename,sep=',')
X=data['airtemp_degc']
```

```python
## Plot hourly temperature data
xindex = (np.arange(0,np.size(X),1))/24.;
v = np.std(X);
labelstr='$\sigma = $' + str(np.round(v)) + '$^{\circ}$C'
plt.plot(xindex, X,label=labelstr);
plt.legend(loc=1,frameon=False)
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('day');
plt.title('Hourly temperature at Truax Field');
```

```python
### Plot hourly temperature data sampled once a day
v = np.std(X[::24]);
labelstr='$\sigma = $' + str(np.round(v)) + '$^{\circ}$C'
plt.plot(xindex[::24],X[::24], color = 'C1',label=labelstr);
plt.legend(loc=1,frameon=False)
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled once-a-day');
```

```python
### Plot hourly temperature data sampled once a month
v = np.std(X[::720]);
labelstr='$\sigma = $' + str(np.round(v)) + '$^{\circ}$C'
plt.plot(xindex[::720],X[::720], color = 'k',label=labelstr);
plt.legend(loc=2,frameon=False)
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled once-a-month');
```

```python
###Plotting all on same plot
plt.plot(xindex, X,label='hourly');
plt.plot(xindex[::24],X[::24], color = 'C1',label='daily');
plt.plot(xindex[::720],X[::720], color = 'k',label='monthly');
plt.legend(loc=2,frameon=False)
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled once-a-month');
```

```python
### Plot sampled daily over one month
v = np.std(X[4000:4000+720:24]);
labelstr='$\sigma = $' + str(np.round(v)) + '$^{\circ}$C'
plt.plot(xindex[4000:4000+720:24],X[4000:4000+720:24], color = 'C2',label=labelstr);
plt.legend(loc=2,frameon=False)
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled daily over one month');
```

```python
### Sampled hourly over one day
v = np.std(X[4000:4000+24:1]);
labelstr='$\sigma = $' + str(np.round(v)) + '$^{\circ}$C'
plt.plot(xindex[4000:4000+24:1],X[4000:4000+24:1], color = 'C3',label=labelstr);
plt.legend(loc=2,frameon=False)
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature ($^{\circ}$C)');
plt.xlabel('days');
plt.title('Sampled hourly over one day');
```
